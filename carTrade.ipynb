```python
from bs4 import BeautifulSoup
import urllib.request
import csv
```

```python
def get_nextpg(city):    
    nextpage = []
    pagelink =  city
    nextpage.append(pagelink)
    for li in nextpage:
        try:
            page2 = urllib.request.urlopen(li)
            soup2 = BeautifulSoup(page2, 'html.parser')
            pagelink =  "https://www.cartrade.com"+soup2.find(class_='next').contents[0].get('href')
            nextpage.append(pagelink)
        except:
            print('Could not find next page')
    scrape_data(nextpage)
```
