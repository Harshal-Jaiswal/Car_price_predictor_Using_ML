```python
from __future__ import print_function

import pandas as pd
pd.__version__
```

```python
df = pd.read_csv(r"C:\Users\T\Downloads\TranslatedDatabase.csv",encoding='ISO-8859-1')
df.describe()
df.head()
```

```python
df=df.drop(['dateCrawled','seller','offerType','abtest','dateCreated','nrOfPictures','postalCode','lastSeen'],axis=1)
```

```python
df.dropna()
```

```python
df = df[df.price != 0 ]
```

```python
df.describe()
```

```python
df=df.dropna()
```

```python
df['vehicleType']=df['vehicleType'].replace(['kleinwagen'], 'mini')
df['vehicleType']=df['vehicleType'].replace(['cabrio'], 'convertible')
df['vehicleType']=df['vehicleType'].replace(['kombi'], 'microbus')
df['vehicleType']=df['vehicleType'].replace(['andere'], 'other')
df['gearbox']=df['gearbox'].replace(['manuell'], 'manual')
df['gearbox']=df['gearbox'].replace(['automatik'], 'automatic')
df['fuelType']=df['fuelType'].replace(['andere'], 'other')
df.price = pd.to_numeric(df.price, errors = 'coerce', downcast= 'integer')
df.yearOfRegistration = pd.to_numeric(df.yearOfRegistration, errors = 'coerce', downcast = 'integer')
df.kilometer = pd.to_numeric(df.kilometer, errors = 'coerce', downcast= 'integer')
df.powerPS = df.powerPS.map( lambda x : df.powerPS.mean() if x == 0 else x )
```

```python
df.describe()
```

```python
df.loc[df['price'].idxmax()]
```

```python
df = df[df.price != 1000000]
```

```python
# for data visualisation and statistical analysis
import numpy as np
from sklearn.cross_validation import train_test_split
import seaborn as sns
sns.set_style("white")
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# here we set the figure size to 15x8
plt.figure(figsize=(15, 8))
# plot two values price per year_model
plt.scatter(df.price, df.yearOfRegistration)
plt.xlabel("price", fontsize=14)
plt.ylabel("year of model", fontsize=14)
plt.title("Scatter plot of price and year of model",fontsize=18)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(15, 12))
sns.stripplot(data = df, x='price', y='brand', jitter=.1)
plt.show()
```

```python

plt.figure(figsize=(17,8))
df.brand.value_counts().nlargest(20).plot(kind='barh')
plt.xlabel('Marks Frequency')
plt.title("Frequency of TOP 20 Brands distribution",fontsize=18)
plt.show()
```

```python
# here we set the figure size to 15x8
plt.figure(figsize=(15, 8))
# plot two values price per year_model
plt.scatter(df.price, df.powerPS, c='orange', marker='x')
plt.xlabel("price", fontsize=14)
plt.ylabel("fiscal power (CV)", fontsize=14)
plt.title("Scatter plot of price and fiscal power",fontsize=18)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(15, 10))
sns.stripplot(data = df, x='fuelType', y='price', jitter=.5)
plt.show()
```

```python
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))
# Compute the correlation matrix
corr = df.drop(['monthOfRegistration'], axis=1).corr()
#print(corr)
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation matrix', 
          fontsize = 20)
plt.show()
```
